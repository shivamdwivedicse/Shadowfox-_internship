# 📊 Air Quality Index (AQI) — Concept Notes

## Delhi Air Quality Analysis Project

---

## 1. AQI Kya Hai?

**Air Quality Index (AQI)** ek single number (0 se 500 tak) hota hai jo batata hai ki
hawa kitni saaf ya kitni khatarnaak hai.

| AQI Range | Category      | Meaning                          |
|-----------|---------------|-----------------------------------|
| 0–50      | Good          | Minimal impact                   |
| 51–100    | Satisfactory  | Minor breathing discomfort       |
| 101–200   | Moderate      | Discomfort to sensitive people   |
| 201–300   | Poor          | Discomfort to most people        |
| 301–400   | Very Poor     | Respiratory illness on exposure  |
| 401–500   | Severe        | Serious health impact, even healthy people |

---

## 2. Problem: Har Pollutant Ka Alag Scale Hota Hai

Raw data mein humare paas 7 pollutants hain: **PM2.5, PM10, NO₂, SO₂, NH₃, CO, O₃**.

Inko seedha compare nahi kar sakte, kyunki:
- PM2.5 = 90 → **already dangerous**
- CO = 90 → **bilkul safe**

Isliye har pollutant ko ek common **0–500 scale** pe convert karna padta hai.
Isko **Sub-Index** kehte hain.

---

## 3. Sub-Index Formula (Linear Interpolation)

Har pollutant ke liye CPCB (Central Pollution Control Board) ne **breakpoint tables**
banayi hain — yeh batati hain ki konsi concentration range, konsi AQI range ke barabar hai.

### Notation:

| Symbol      | Meaning                                       |
|-------------|------------------------------------------------|
| $C_p$       | Measured concentration of pollutant             |
| $C_{low}$   | Lower bound of the concentration bracket        |
| $C_{high}$  | Upper bound of the concentration bracket        |
| $I_{low}$   | AQI value corresponding to $C_{low}$            |
| $I_{high}$  | AQI value corresponding to $C_{high}$           |
| $I_p$       | Sub-index (final answer) for that pollutant     |

### Formula:

$$
I_p = I_{low} + \frac{I_{high} - I_{low}}{C_{high} - C_{low}} \times (C_p - C_{low})
$$

**Simple words mein:** "Yeh concentration apne bracket ke andar kitne % pe hai,
utna hi % AQI range ke andar bhi le lo."

### Example:
PM2.5 = 75, jo 61–90 bracket (AQI 101–200) mein aata hai:

$$
I_p = 101 + \frac{200-101}{90-61} \times (75-61) ≈ 149
$$

---

## 4. Overall AQI Kaise Nikalte Hain

Ek din ka final AQI = **sabse zyada wala sub-index (MAX)**, average nahi.

$$
AQI_{final} = \max(I_{PM2.5},\ I_{PM10},\ I_{NO_2},\ I_{SO_2},\ I_{NH_3},\ I_{CO},\ I_{O_3})
$$

**Kyun max, average nahi?**
Agar ek gas bhi bahut zyada khatarnaak hai, poori hawa khatarnaak maani jaati hai —
chahe baaki gases normal hi kyun na hon.

Jo pollutant yeh max value deta hai, use **Dominant Pollutant** kehte hain.

---

## 5. Averaging Rule (Important!)

CPCB AQI **raw hourly values** pe nahi, balki **averaged values** pe calculate hota hai:

| Pollutant Group                  | Averaging Period |
|-----------------------------------|-------------------|
| PM2.5, PM10, NO₂, SO₂, NH₃         | 24-hour average   |
| CO, O₃                            | 8-hour average (simplified to 24-hr average in this project) |

Isliye hourly data ko pehle **daily average** mein convert kiya gaya
(`groupby('full_date')`), tabhi AQI nikala gaya.

### See the code now

```python
def get_subindex(conc,breakpoints):
    for (c_low , c_high , i_low , i_high) in breakpoints:
        if c_low <= conc <=c_high:
            return i_low +(i_high - i_low) / (c_high - c_low)*(conc - c_low)
    # if concentration exceeds the highest breakpoint, extrapolate linearly
    last = breakpoints[-1]
    if conc > last[1]:
        return last[3] + (conc - last[1]) * (last[3]-last[2])/(last[1]-last[0])
    return None
```

```python
daily = df.groupby('full_date').agg({
    'pm2_5': 'mean',
    'pm10': 'mean',
    'nitrogen_dioxide': 'mean',
    'sulphur_dioxide': 'mean',
    'ammonia': 'mean',
    'co_mg': 'mean',
    'ozone': 'mean'
}).reset_index()
```

```python
print(daily.shape)
```

## calculating AQI

```python
daily['si_pm2_5'] = daily['pm2_5'].apply(lambda x: get_subindex(x, pm2_5_bp))
daily['si_pm10']  = daily['pm10'].apply(lambda x: get_subindex(x, pm10_bp))
daily['si_no2']   = daily['nitrogen_dioxide'].apply(lambda x: get_subindex(x, no2_bp))
daily['si_so2']   = daily['sulphur_dioxide'].apply(lambda x: get_subindex(x, so2_bp))
daily['si_nh3']   = daily['ammonia'].apply(lambda x: get_subindex(x, nh3_bp))
daily['si_co']    = daily['co_mg'].apply(lambda x: get_subindex(x, co_bp))
daily['si_o3']    = daily['ozone'].apply(lambda x: get_subindex(x, o3_bp))

si_cols = ['si_pm2_5','si_pm10','si_no2','si_so2','si_nh3','si_co','si_o3']

daily['AQI'] = daily[si_cols].max(axis=1)
daily['dominant_pollutant'] = daily[si_cols].idxmax(axis=1).str.replace('si_', '')

print(daily[['full_date','AQI','dominant_pollutant']].to_string())
```

#### Distribution of AQI categories — What proportion of the observed period falls into each AQI category (Good/Satisfactory/Moderate/Poor/Very Poor/Severe)?

```python
def categorize(aqi):
    if aqi <= 50: return 'Good'
    elif aqi <= 100: return 'Satisfactory'
    elif aqi <= 200: return 'Moderate'
    elif aqi <= 300: return 'Poor'
    elif aqi <= 400: return 'Very Poor'
    else: return 'Severe'

daily['category'] = daily['AQI'].apply(categorize)
print(daily['category'].value_counts())
```

## Average Pollutant Levels vs Safe Limits

```python
safe_limits = {
    'pm2_5':60, 
    'pm10':100, 
    'nitrogen_dioxide':80, 
    'sulphur_dioxide':80, 
    'ammonia':400, 
    'carbon_monoxide':2000, # µg/m3 equivalent
    'ozone':100
}

pollutants_cols = ['pm2_5','pm10','nitrogen_dioxide','sulphur_dioxide','ammonia','carbon_monoxide','ozone'] 
avg_levels = df[pollutants_cols].mean()

comparison_df = pd.DataFrame({
    'Observed_Average' : avg_levels, 
    'Safe_limit' : pd.Series(safe_limits)
})
comparison_df['Time_Exceeded'] = comparison_df['Observed_Average']/comparison_df['Safe_limit']
print(comparison_df)
```

 ## Outliers (Abnormal Spikes)

```python
plt.figure(figsize=(10,6))
df.boxplot(column=['pm2_5','pm10','carbon_monoxide'], rot=0)
plt.title('Q9: Outlier Detection - Boxplot')
plt.ylabel('Concentration')
plt.tight_layout()
plt.savefig('q9_boxplot.png', dpi=150)
plt.show()
```

```python
def find_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    return outliers[['date', column]]

pm25_outliers = find_outliers('pm2_5')
print(pm25_outliers)
print(f"Total outlier hours: {len(pm25_outliers)}")
```
