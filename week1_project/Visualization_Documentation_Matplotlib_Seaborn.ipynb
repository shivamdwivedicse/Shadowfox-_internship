# Introduction

Data Visualization is the process of representing data in graphical form. It helps in identifying trends, patterns, relationships, and outliers within a dataset.

In Data Science, visualization is one of the most important steps of Exploratory Data Analysis (EDA). Instead of reading thousands of rows of data, graphs allow us to understand information quickly and make better decisions.

Python provides several powerful visualization libraries. In this documentation, we will explore Matplotlib and Seaborn.

# Dataset Used

**Dataset Name:** Iris Dataset

The Iris dataset is a famous multivariate dataset introduced by Ronald Fisher. It contains 150 observations of iris flowers from three different species: Setosa, Versicolor, and Virginica.

### Features
- Sepal Length (cm).
- Sepal Width (cm).
- Petal Length (cm).
- Petal Width (cm).
- Species

The dataset is widely used for learning Data Science, Machine Learning, and Data Visualization because it is clean, balanced, and easy to understand.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn theme
sns.set_theme(style="whitegrid")
```

# Library 1 : Matplotlib

## What is Matplotlib?

Matplotlib is one of the most widely used Python libraries for data visualization. It provides complete control over plots and is suitable for creating publication-quality graphs.

### Features

- Easy to learn.
- Highly customizable.
- Supports 2D plotting.
- Works well with NumPy and Pandas.
- Industry standard visualization library.

### Common Applications

- Exploratory Data Analysis.
- Business Reporting.
- Scientific Research.
- Machine Learning.
- Dashboard Development.

```python
# Load Dataset
iris = sns.load_dataset("iris")

# Display first 5 rows
iris.head()
```

## 1. Line Plot

### What is a Line Plot?

A Line Plot is used to visualize trends or changes over continuous data. It connects data points with straight lines.

**Use Cases**
- Sales Trend
- Temperature Change
- Stock Price Analysis
- Revenue Growth

```python
plt.figure(figsize=(8,5))

plt.plot(
    iris.index,
    iris["sepal_length"],
    color="blue",
    linewidth=2,
    label="Sepal Length"
)

plt.title("Sepal Length Trend")
plt.xlabel("Observation")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.grid(True)

plt.show()
```

## 2. Scatter Plot

### What is a Scatter Plot?

A Scatter Plot shows the relationship between two numerical variables.

**Use Cases**
- Correlation Analysis
- Outlier Detection
- Feature Relationship
- Machine Learning EDA

```python
plt.figure(figsize=(8,5))

plt.scatter(
    iris["sepal_length"],
    iris["petal_length"],
    color="red"
)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.show()
```

## 3. Bar Chart

### What is a Bar Chart?

A Bar Chart compares values across different categories.

**Use Cases**
- Category Comparison
- Sales by Product
- Student Marks
- Population Comparison

```python
species_mean = iris.groupby("species")["sepal_length"].mean()

plt.figure(figsize=(8,5))

plt.bar(
    species_mean.index,
    species_mean.values
)

plt.title("Average Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")

plt.show()
```

## 4. Histogram

### What is a Histogram?

A Histogram shows the frequency distribution of numerical data.

**Use Cases**
- Data Distribution
- Detect Skewness
- Identify Outliers
- Statistical Analysis

```python
plt.figure(figsize=(8,5))

plt.hist(
    iris["sepal_length"],
    bins=10,
    edgecolor="black"
)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()
```

## 5. Pie Chart

### What is a Pie Chart?

A Pie Chart represents the proportion of different categories as slices of a circle.

**Use Cases**
- Market Share
- Budget Distribution
- Population Percentage
- Survey Results

```python
species_count = iris["species"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    species_count,
    labels=species_count.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Species Distribution")

plt.show()
```

## 6. Box Plot

### What is a Box Plot?

A Box Plot summarizes the distribution of data using quartiles and helps identify outliers.

**Use Cases**
- Outlier Detection
- Data Spread
- Statistical Analysis

```python
plt.figure(figsize=(8,5))

plt.boxplot(iris["sepal_length"])

plt.title("Box Plot of Sepal Length")

plt.ylabel("Sepal Length")

plt.show()
```

## 7. Area Plot

### What is an Area Plot?

An Area Plot is similar to a line plot but fills the area below the line.

**Use Cases**
- Trend Analysis
- Cumulative Growth
- Time Series

```python
plt.figure(figsize=(8,5))

plt.fill_between(
    iris.index,
    iris["sepal_length"],
    alpha=0.5
)

plt.plot(iris.index, iris["sepal_length"])

plt.title("Area Plot of Sepal Length")

plt.xlabel("Observation")

plt.ylabel("Sepal Length")

plt.show()
```

## 8. Stem Plot

### What is a Stem Plot?

A Stem Plot displays data points as vertical lines with markers.

**Use Cases**
- Signal Processing
- Discrete Data
- Frequency Analysis

```python
plt.figure(figsize=(10,5))

plt.stem(iris.index[:30], iris["sepal_length"][:30])

plt.title("Stem Plot")

plt.xlabel("Observation")

plt.ylabel("Sepal Length")

plt.show()
```

## 9. Subplots

### What are Subplots?

Subplots allow multiple graphs to be displayed in a single figure.

**Use Cases**
- Dashboard Creation
- Multiple Comparisons
- Exploratory Data Analysis

```python
fig, ax = plt.subplots(1,2, figsize=(12,5))

ax[0].hist(iris["sepal_length"])
ax[0].set_title("Histogram")

ax[1].boxplot(iris["sepal_length"])
ax[1].set_title("Box Plot")

plt.tight_layout()

plt.show()
```

## 10. Save Figure

### What is Save Figure?

Matplotlib allows plots to be saved in different formats such as PNG, JPG, and PDF for reports and presentations.

**Use Cases**
- Reports
- Research Papers
- Dashboards
- Presentations

```python
plt.figure(figsize=(8,5))

plt.plot(iris.index, iris["petal_length"])

plt.title("Petal Length Trend")

plt.savefig("petal_length_plot.png", dpi=300)

plt.show()
```

# Advantages of Matplotlib

- Easy to use.
- Highly customizable.
- Wide variety of plots.
- Excellent documentation.
- Integrates with NumPy and Pandas.
- Industry standard visualization library.

# Limitations of Matplotlib

- Requires more code than Seaborn.
- Default style is basic.
- Statistical plots require extra effort.
- Interactive visualizations are limited.

# Conclusion

Matplotlib is one of the most powerful and widely used visualization libraries in Python. It offers complete control over graphs and is suitable for creating publication-quality visualizations. It is commonly used in data analysis, machine learning, scientific research, and business analytics.

# Library 2 : Seaborn

## What is Seaborn?

Seaborn is a high-level Python data visualization library built on top of Matplotlib. It simplifies the process of creating attractive and informative statistical graphics with minimal code.

Unlike Matplotlib, Seaborn provides beautiful default themes, built-in datasets, and advanced statistical plotting capabilities. It is widely used for Exploratory Data Analysis (EDA), Machine Learning, and Data Science projects.

### Features

- Built on top of Matplotlib.
- Beautiful default themes and color palettes.
- Easy to create statistical plots.
- Works seamlessly with Pandas DataFrames.
- Built-in datasets for learning and analysis.
- Less code with better visualization.

### Common Applications

- Exploratory Data Analysis (EDA).
- Machine Learning Projects.
- Statistical Data Visualization.
- Business Analytics.
- Correlation Analysis.
- Data Distribution Analysis.

# 1. Scatter Plot

### What is a Scatter Plot?

A Scatter Plot is used to visualize the relationship between two numerical variables.

**Use Cases**
- Correlation Analysis
- Pattern Detection
- Outlier Identification

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species"
)

plt.title("Scatter Plot")
plt.show()
```

# 2. Line Plot

### What is a Line Plot?

A Line Plot displays trends or changes in data over a continuous variable.

**Use Cases**
- Sales Trend
- Stock Prices
- Time Series Analysis

```python
plt.figure(figsize=(8,5))

sns.lineplot(
    data=iris,
    x=iris.index,
    y="sepal_length"
)

plt.title("Line Plot")
plt.show()
```

# 3. Bar Plot

### What is a Bar Plot?

A Bar Plot compares numerical values across different categories.

**Use Cases**
- Category Comparison
- Average Analysis

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=iris,
    x="species",
    y="sepal_length"
)

plt.title("Bar Plot")
plt.show()
```

# 4. Count Plot

### What is a Count Plot?

A Count Plot shows the frequency of each category.

**Use Cases**
- Category Distribution
- Class Balance

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=iris,
    x="species"
)

plt.title("Count Plot")
plt.show()
```

# 5. Histogram

### What is a Histogram?

A Histogram displays the frequency distribution of numerical values.

**Use Cases**
- Data Distribution
- Outlier Detection

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=iris,
    x="petal_length",
    bins=15,
    kde=True
)

plt.title("Histogram")
plt.show()
```

# 6. Box Plot

### What is a Box Plot?

A Box Plot summarizes data distribution and highlights outliers.

**Use Cases**
- Outlier Detection
- Distribution Analysis

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=iris,
    x="species",
    y="petal_length"
)

plt.title("Box Plot")
plt.show()
```

# 7. Violin Plot

### What is a Violin Plot?

A Violin Plot combines a Box Plot with a density plot.

**Use Cases**
- Distribution Comparison
- Statistical Analysis

```python
plt.figure(figsize=(8,5))

sns.violinplot(
    data=iris,
    x="species",
    y="petal_length"
)

plt.title("Violin Plot")
plt.show()
```

# 8. Heatmap

### What is a Heatmap?

A Heatmap visualizes correlations between numerical variables using colors.

**Use Cases**
- Correlation Analysis
- Feature Selection

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    iris.select_dtypes(include="number").corr(),
    annot=True,
    cmap="Blues"
)

plt.title("Correlation Heatmap")
plt.show()
```

# 9. Pair Plot

### What is a Pair Plot?

A Pair Plot displays pairwise relationships between numerical variables.

**Use Cases**
- Exploratory Data Analysis
- Feature Relationship Analysis

```python
sns.pairplot(
    iris,
    hue="species"
)

plt.show()
```

# Advantages of Seaborn

- Beautiful default themes.
- Less code than Matplotlib.
- Built-in datasets.
- Excellent statistical visualizations.
- Easy integration with Pandas.

# Limitations of Seaborn

- Less customizable than Matplotlib.
- Depends on Matplotlib.
- Limited support for highly customized plots.

# Conclusion

Seaborn is a high-level visualization library built on Matplotlib. It simplifies the creation of attractive statistical graphics and is widely used for Exploratory Data Analysis (EDA), Machine Learning, and Data Science projects.

# Matplotlib vs Seaborn

| Feature | Matplotlib | Seaborn |
|----------|------------|----------|
| Level | Low-Level Library | High-Level Library |
| Built On | Standalone | Built on Matplotlib |
| Ease of Use | Moderate | Easy |
| Customization | High | Moderate |

| Code Length | More | Less |
| Best For | Custom Graphs | Exploratory Data Analysis (EDA) |
